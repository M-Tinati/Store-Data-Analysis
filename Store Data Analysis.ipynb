```python
from faker import Faker
import pandas as pd
from IPython.display import display
from datetime import date
import matplotlib.pyplot as plt

Faker.seed(101) #    
fake = Faker()

#  
names = [] 
ages = []
genders  = []
PaymentMethods = []
MembershipStatuss = []

for _ in range(13):
    name = fake.name()
    names.append(name)  
    
    # 
    birth_date = fake.date_of_birth(minimum_age=18, maximum_age=70)
    today = date.today()
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    ages.append(age)
    
    # 
    gender = fake.random_element(elements=('Male', 'Female'))
    genders.append(gender)
    
    # 
    PaymentMethod = fake.random_element(elements=('cash', 'credit card'))
    PaymentMethods.append(PaymentMethod)
    
    # 
    MembershipStatus = fake.random_element(elements=('vip' , 'simpel'))
    MembershipStatuss.append(MembershipStatus)

data = {
    "name": names, 
    "age" : ages,
    "gender" : genders,
    "PaymentMethod" : PaymentMethods,
    "Membership Status" : MembershipStatuss
}

df = pd.DataFrame(data)
#pd

# 
PaymentMethod_labels = df['PaymentMethod'].value_counts()
plt.pie(PaymentMethod_labels , labels=PaymentMethod_labels.index, autopct='%1.2f%%', startangle=25)
plt.title('PaymentMethod')
plt.show()

MembershipStatus_labels = df['Membership Status'].value_counts()
plt.pie(MembershipStatus_labels , labels=MembershipStatus_labels.index, autopct='%1.2f%%', startangle=25)
plt.title('MembershipStatus')
plt.show()

age_labels = df['age'].value_counts()
plt.pie(age_labels , labels=age_labels.index, autopct='%1.2f%%', startangle=25)
plt.title('age')
plt.show()

gender_labels = df['gender' ].value_counts()
plt.pie(gender_labels, labels=gender_labels.index, autopct='%1.1f%%', startangle=30)
plt.title('Genders')
plt.show()
```
